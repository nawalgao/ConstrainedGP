```python
def rbf(x1, x2, var, l):
    diff = x1 - x2.T
    sqr = diff**2
    val = var*np.exp(-1./(2*l**2)*sqr)
    return val

def rbf_d(x1, x2, var, l):
    diff = x1 - x2.T
    sqr = diff**2
    val = var/l**2*np.exp(-1./(2*l**2)*sqr)*(1 - 1./l**2*sqr)
    return val

def rbf_n_d(x1, x2, var, l):
    """
    cov between GP at x1 points and derivative of GP at x2 points
    """
    diff = x1 - x2.T
    sqr = diff**2
    val = -var/l**2*np.exp(-1./(2*l**2)*sqr)*diff
    return val

def K_m_m(X, X_prime, var, l):
    k1 = rbf(X, X, var, l)
    k2 = rbf_n_d(X, X_prime, var, l)
    k3 = k2.T
    k4 = rbf_d(X_prime, X_prime, var, l)
    K11 = np.hstack([k1,k2])
    K22 = np.hstack([k3, k4])
    KK = np.vstack([K11, K22])
    return KK

def K_m_n(Xnew, X, X_prime, var, l):
    k1 = rbf(X, Xnew, var, l)
    k2 = rbf_n_d(Xnew, X_prime, var, l)
    kk = np.vstack([k1, k2.T])
    return kk
    
    
```
