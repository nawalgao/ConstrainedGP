# Expected Utility Improvement

```python
exp_utility_max = np.max(mtrainmat, axis = 1)
exp_utility_max_ind = np.argmax(mtrainmat, axis = 1)
train_max_ut_state = X[:,1][exp_utility_max_ind]
```

```python
def EUI(X_norm, Y, config_file):
    m, samples = pref_model_train.train(X_norm, Y, config_file)
    mtrainmat, vartrainmat, mgridmat, vargridmat = pref_model_predict.u(X_norm, Xgridnorm, samples, m)
    exp_utility_max = np.max(mtrainmat, axis = 1)
    mdiff = mgridmat - exp_utility_max[:,None]
    siggridmat = np.sqrt(vargridmat)
    Z = mdiff/siggridmat
    pdf = scipy.stats.norm.pdf(Z)
    cdf = scipy.stats.norm.cdf(Z)
    exp_imp = mdiff*cdf + siggridmat*pdf
    mean_exp_imp = np.mean(exp_imp, axis = 0)
    max_exp_imp_ind = np.argmax(mean_exp_imp)
    next_state = Xgrid[max_exp_imp_ind]
    #print next_state
    return next_state
```

```python
n = EUI(X_norm, Y, config_file)
```
